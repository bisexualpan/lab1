#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
# Your code goes here

def count_vowels(s: str) -> int:
    vowels = set('aeiou')
    count = 0 
    for c in s.lower():
        if c in vowels:
            count += 1
    return count  
```

```python
%%ipytest

# ── Тест 1: пустая строка ─────────────────────────────────────────────────
def test_empty():
    assert count_vowels("") == 0

# ── Тест 2: только согласные ───────────────────────────────────────────
def test_consonants_only():
    assert count_vowels("brtdcdfghtklmnq") == 0

# ── Тест 3: только гласные ────────────────────────────────────────────────
def test_vowels_only():
    assert count_vowels("aeiou") == 5

# ── Тест 4: рандомный регистр ─────────────────────────────────────────────
def test_mixed_case():
    assert count_vowels("aBcDEfFGHKLMiO") == 4

# ── Тест 5: с пробелами ────
def test_with_spaces():
    assert count_vowels(" urfu sucks ") == 3

# ── Тест 6: повторяющиеся гласные ─────────────────────────────────────────
def test_repeating_vowels():
    assert count_vowels("aaaeeeiiiooouuu") == 15
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
# Your code goes here

def check_unique(s: str) -> bool:
    return len(s) == len(set(s)) # сет чтобы уникальные вытащить ... вот так вот
```

```python
%%ipytest

# ── Тест 1: пустая ─────────────────────────────────────────────────
def test_empty():
    assert check_unique("") is True

# ── Тест 2: все уникальные ─────────────────────────────────────────
def test_true():
    assert check_unique("qwerty") is True
    assert check_unique("123456") is True
    assert check_unique("Cc") is True

# ── Тест 3: есть повторяющиеся  ────────────────────────────────────
def test_false():
    assert check_unique("nigger") is False
    assert check_unique("lol") is False
    assert check_unique("assertttion") is False

# ── Тест 4: с пробелами (пробел тоже учитывается) ────────────────────────────
def test_with_spaces():
    assert check_unique("q w") is True
    assert check_unique("q  w") is False
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits (a: int) -> int:
    return bin(a).count('1')
```

```python
%%ipytest

# ── Тест 1: ноль ──────────────────────────────────────────────────────────
def test_zero():
    assert count_bits(0) == 0

# ── Тест 2: единица ───────────────────────────────────────────────────────
def test_one():
    assert count_bits(1) == 1

# ── Тест 3: степени двойки ────────────────────────────────────────────────
def test_powers_of_two():
    assert count_bits(2) == 1
    assert count_bits(4) == 1
    assert count_bits(8) == 1
    assert count_bits(16) == 1

# ── Тест 4: произвольные числа ─────────────────────────────────────────────
def test_examples():
    assert count_bits(5) == 2   # 101
    assert count_bits(7) == 3   # 111
    assert count_bits(10) == 2  # 1010
    assert count_bits(255) == 8

# ── Тест 5: большое число ─────────────────────────────────────────────────
def test_large():
    assert count_bits(2**20 - 1) == 20
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n: int) -> int:
    steps = 0
    while n > 9:  # пока число больше 9 перемножаем цифры
        p = 1
        for digit in str(n):
            p = p * int(digit)
        n = p
        steps = steps + 1
    
    return steps
```

```python
%%ipytest

# ── Тест 1: однозначное ─────────────────────────────────────────────
def test_single_digit():
    assert magic(4) == 0
    assert magic(0) == 0
    assert magic(9) == 0

# ── Тест 2: пример из условия (39) ─────────────────────────────────────────────────────
def test_example_39():
    assert magic(39) == 3

# ── Тест 3: пример из условия (999) ────────────────────────────────────────────────────
def test_example_999():
    assert magic(999) == 4

# ── Тест 4: число 25 ──────────────────────────────────────────────────────
def test_25():
    assert magic(25) == 2   # 2*5=10 → 1*0=0

# ── Тест 5: число 10 ──────────────────────────────────────────────────────
def test_10():
    assert magic(10) == 1

# ── Тест 6: число 88 ──────────────────────────────────────────────────────
def test_88():
    assert magic(88) == 3   # 88→64→24→8
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    # Your code goes here
    pred = np.asarray(pred)
    true = np.asarray(true)
    return np.sqrt(np.mean((pred - true) ** 2))
```

```python
%%ipytest

# ── Тест 1: идентичные векторы ────────────────────────────────────────────
def test_identical_vectors():
    a = [1, 2, 3]
    b = [1, 2, 3]
    np.testing.assert_almost_equal(mse(a, b), 0.0)

# ── Тест 2: постоянная разность ───────────────────────────────────────────
def test_constant_difference():
    a = [1, 2, 3]
    b = [4, 5, 6]   # разность 3 для каждого
    np.testing.assert_almost_equal(mse(a, b), 3.0)

# ── Тест 3: один элемент ──────────────────────────────────────────────────
def test_single_element():
    np.testing.assert_almost_equal(mse([5], [3]), 2.0)

# ── Тест 4: отрицательные значения ────────────────────────────────────────
def test_negative_values():
    a = [-1, -2]
    b = [1, 2]
    np.testing.assert_almost_equal(mse(a, b), np.sqrt(10))

# ── Тест 5: работа с numpy-массивами ──────────────────────────────────────
def test_numpy_arrays():
    a = np.array([1.5, 2.5])
    b = np.array([3.5, 4.5])
    np.testing.assert_almost_equal(mse(a, b), 2.0)
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n: int) -> str:
    if n <= 1:
        return ""
    
    result = []
    temp = n
    divisor = 2
    
    while divisor * divisor <= temp:
        if temp % divisor == 0:
            count = 0
            while temp % divisor == 0:
                temp //= divisor
                count += 1
            if count == 1:
                result.append(f"({divisor})")
            else:
                result.append(f"({divisor}**{count})")
        
        divisor += 1 if divisor == 2 else 2
    
    if temp > 1:
        result.append(f"({temp})")
    
    return "".join(result)
```

```python
%%ipytest

# ── Тест 1: число 1 ───────────────────────────────────────────────────────
def test_one():
    assert prime_factors(1) == ""

# ── Тест 2: простое число ─────────────────────────────────────────────────
def test_prime():
    assert prime_factors(13) == "(13)"

# ── Тест 3: степень простого ──────────────────────────────────────────────
def test_power():
    assert prime_factors(8) == "(2**3)"
    assert prime_factors(9) == "(3**2)"

# ── Тест 4: пример 86240 ──────────────────────────────────────────────────
def test_example():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"

# ── Тест 5: число 60 ──────────────────────────────────────────────────────
def test_60():
    assert prime_factors(60) == "(2**2)(3)(5)"

# ── Тест 6: число 100 ─────────────────────────────────────────────────────
def test_100():
    assert prime_factors(100) == "(2**2)(5**2)"
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number: int):
    
    if number < 1:
        return "It is impossible"
    total = 0
    k = 0
    while total < number:
        k += 1
        total += k * k
    return k if total == number else "It is impossible"
```

```python
%%ipytest
# ── Тест 1: n = 1 (1²) ────────────────────────────────────────────────────
def test_1():
    assert pyramid(1) == 1

# ── Тест 2: n = 5 (1² + 2²) ───────────────────────────────────────────────
def test_5():
    assert pyramid(5) == 2

# ── Тест 3: n = 14 (1² + 2² + 3²) ─────────────────────────────────────────
def test_14():
    assert pyramid(14) == 3

# ── Тест 4: n = 30 (1² + 2² + 3² + 4²) ────────────────────────────────────
def test_30():
    assert pyramid(30) == 4

# ── Тест 5: n = 55 (1² + ... + 5²) ────────────────────────────────────────
def test_55():
    assert pyramid(55) == 5

# ── Тест 6: невозможно (число 2) ──────────────────────────────────────────
def test_impossible_2():
    assert pyramid(2) == "It is impossible"

# ── Тест 7: невозможно (число 3) ──────────────────────────────────────────
def test_impossible_3():
    assert pyramid(3) == "It is impossible"

# ── Тест 8: невозможно (число 6) ──────────────────────────────────────────
def test_impossible_6():
    assert pyramid(6) == "It is impossible"

# ── Тест 9: невозможно (отрицательное число) ──────────────────────────────
def test_negative():
    assert pyramid(-5) == "It is impossible"

# ── Тест 10: невозможно (ноль) ────────────────────────────────────────────
def test_zero():
    assert pyramid(0) == "It is impossible"
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number: int) -> bool:
    '''
    Определяет, является ли число сбалансированным:
    для нечётной длины средняя цифра игнорируется,
    для чётной — две средние цифры игнорируются.
    Сумма цифр слева от средних равна сумме справа.
    '''
    s = str(number)
    n = len(s)
    mid = n // 2
    if n % 2 == 0:
        left_sum = sum(int(ch) for ch in s[:mid-1])
        right_sum = sum(int(ch) for ch in s[mid+1:])
    else:
        left_sum = sum(int(ch) for ch in s[:mid])
        right_sum = sum(int(ch) for ch in s[mid+1:])
    return left_sum == right_sum
```

```python
%%ipytest

# ── Тест 1: нечётная длина, сбалансировано ────────────────────────────────
def test_odd_balanced():
    assert is_balanced(12321) is True   # 1+2 == 2+1
    assert is_balanced(121) is True     # 1 == 1
    assert is_balanced(5) is True       # пустые суммы 0==0

# ── Тест 2: нечётная длина, несбалансировано ──────────────────────────────
def test_odd_unbalanced():
    assert is_balanced(12341) is False  # 1+2=3, 4+1=5

# ── Тест 3: чётная длина, сбалансировано ──────────────────────────────────
def test_even_balanced():
    assert is_balanced(1221) is True    # 1 == 1
    assert is_balanced(12344321) is True # 1+2+3 == 3+2+1
    assert is_balanced(12) is True      # пустые суммы 0==0

# ── Тест 4: чётная длина, несбалансировано ────────────────────────────────
def test_even_unbalanced():
    assert is_balanced(1234) is False   # 1 != 4
    assert is_balanced(1230) is False

# ── Тест 5: однозначные числа ─────────────────────────────────────────────
def test_single_digit():
    assert is_balanced(0) is True
    assert is_balanced(9) is True
```
